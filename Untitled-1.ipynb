```python
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import json

with open("data/merged_cleaned.xls", "r", encoding="utf-8", errors="replace") as f:
    df_merged = pd.read_csv(f)

# Convert 'Installation Date' to datetime and extract the year as string
df_merged['Installation Date'] = pd.to_datetime(df_merged['Installation Date'], errors='coerce')
df_merged['Installation Year'] = df_merged['Installation Date'].dt.year.astype(str)
df_merged['Installation Year'] = pd.to_numeric(df_merged['Installation Year'], errors='coerce')
  # Convert to string

# Load the simplified GeoJSON for Nepal's districts
with open('data/nepal-districts-new-reduced.json') as geo:
    nepal_districts = json.load(geo)

# Extract province code and add it to the dataframe
province_mapping = {feature['properties']['DIST_PCODE']: feature['properties']['ADM1_PCODE']
                    for feature in nepal_districts['features']}
df_merged['Province Code'] = df_merged['DISTRICT_PCODE'].map(province_mapping)

# Convert 'DIST_PCODE' to string and handle NaN values
df_merged['DISTRICT_PCODE'] = df_merged['DISTRICT_PCODE'].fillna('Unknown').astype(str)

# Province names mapped to their codes
province_names = {
    'NP01': 'Koshi', 'NP02': 'Madhesh', 'NP03': 'Bagmati', 'NP04': 'Gandaki',
    'NP05': 'Lumbini', 'NP06': 'Karnali', 'NP07': 'Sudurpashchim'
}

# Add Province Name column by mapping from Province Code
df_merged['Province Name'] = df_merged['Province Code'].map(province_names)

# Save the updated DataFrame to CSV
df_merged.to_csv("data/merged_cleaned.xls", index=False)
```

```python
import pandas as pd

# Re-read the original file
df_original = pd.read_csv('data/merged_cleaned.xls')  # Replace with the original file

# Recompute 'District Count' and 'Total District Count' columns based on groupings (if needed)
df_original['District Count_x'] = df_original.groupby('DISTRICT_PCODE')['District'].transform('count')
df_original['District Count_y'] = df_original.groupby('DISTRICT_PCODE')['District'].transform('count')
df_original['Total District Count_x'] = df_original.groupby('DISTRICT_PCODE')['Total District Count'].transform('sum')
df_original['Total District Count_y'] = df_original.groupby('DISTRICT_PCODE')['Total District Count'].transform('sum')

# Merge back with your modified DataFrame (assuming you kept other columns)
df_modified = pd.read_csv('data/merged_cleaned.xls')  # The modified file after dropping columns

# Merge the data on the common column(s), like 'DISTRICT_PCODE' or 'District', etc.
df_restored = pd.merge(df_modified, df_original[['DISTRICT_PCODE', 'District Count_x', 'District Count_y', 'Total District Count_x', 'Total District Count_y']], on='DISTRICT_PCODE', how='left')

# Save the restored DataFrame to a new file
df_restored.to_csv('data/merged_cleaned.xls', index=False)
```
